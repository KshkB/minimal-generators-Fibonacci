# Minimal Fibonacci Generators

Given a number $N$, what are the minimimal, positive integer pair $(i, j)$ such that the Fibonacci sequence generated by $(i, j)$ contains $N$?

**Note.** Tuples (here, integral 2-vectors) are ordered by their internal Pythagorean length. That is, treat $(i, j)$ as lengths of legs of a right-angle triangle. The *internal Pythagorean length* of the tuple $(i, j)$ is then $\|(i, j)\| = \sqrt{i^2 + j^2}$. Accordingly, the ordering on tuples is given by: 

\begin{align}
(i, j) \leq (i^\prime, j^\prime)&&\mbox{iff}&&\|(i, j)\| \leq \|(i^\prime, j^\prime)\|
\end{align}
